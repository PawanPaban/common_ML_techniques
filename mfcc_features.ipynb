```python
import librosa as lb
import os

import numpy as np
import matplotlib.pyplot as plt
import pandas as p
```

```python
metadata = p.read_csv("H:\Second_DiCOVA_Challenge_Dev_Data_Release\Second_DiCOVA_Challenge_Dev_Data_Release\metadata.csv",delimiter = ' ')
metadata.values
```

```python
x, sr = lb.load("H:\Second_DiCOVA_Challenge_Dev_Data_Release\Second_DiCOVA_Challenge_Dev_Data_Release\AUDIO\cough\ABVXITDD.flac", sr =44100)
mfcc = lb.feature.mfcc(x,sr=sr,hop_length=512,n_mfcc=15)

#to load multiple audio files
audio_fpath = "H://Second_DiCOVA_Challenge_Dev_Data_Release//Second_DiCOVA_Challenge_Dev_Data_Release//AUDIO//cough//"
audio_clips = os.listdir(audio_fpath)
print("No. of .flac files in audio\cough folder = ",len(audio_clips))
m=np.ndarray(965,)
for i in range(965):
    y, sr = lb.load(audio_fpath + audio_clips[i], sr =44100)
type(y)    
```

```python
%matplotlib inline
#visualising mfcc
plt.figure(figsize=(10,10))
lb.display.specshow(mfcc,x_axis="time",sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

```python
#first and second derivatives feature of mfcc w.r.t. time
delta_mfcc=lb.feature.delta(mfcc)
delta2_mfcc=lb.feature.delta(mfcc, order = 2)
```

```python
plt.figure(figsize=(10,10))
lb.display.specshow(delta_mfcc,x_axis="time",sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

```python
plt.figure(figsize=(10,5))
lb.display.specshow(delta2_mfcc,x_axis="time",sr=sr)
plt.colorbar(format="%+2f")
plt.show()
```

```python
concatenated_mfccs = np.concatenate([mfcc,delta_mfcc,delta2_mfcc])
concatenated_mfccs.shape
```
