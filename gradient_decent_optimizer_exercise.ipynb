```python
import pandas as p
stuff = p.read_csv("H:/py-master/py-master/ML/3_gradient_descent/Exercise/test_scores.csv")
stuff
```

```python
tuff = stuff.drop('name',axis='columns')
tuff
```

```python
iteration = 100000
m,b=0,0
learning_rate = 0.0001
for i in range(iteration):
    y = tuff.cs
    x = tuff.math
    n = len(x)
    yp = m * x + b
    cost = (1/n)*sum(val**2 for val in (y-yp))
    c_md = -(2/n)*sum(val for val in (x*(y-yp)))
    c_bd = -(2/n)*sum(val for val in (y-yp))
    m = m - learning_rate*c_md
    b = b - learning_rate*c_bd
    print(f"m = {m},b = {b}, cost = {cost}")
```
